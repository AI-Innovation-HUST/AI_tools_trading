```python
# h = h
# l = low
# o = o
# c = c
# v = vol

indicators = ta.get_functions()

indicators.remove('AD')
indicators.remove('ADOSC')
indicators.remove('MFI')
indicators.remove('OBV')

hlcv_indicators = ['AD','ADOSC','MFI']
cl_indicators = ['OBV']


data_full_with_indi = []




# data_full = data_full.copy()
for indicator in indicators:
    print(indicator)        # indi need 1 value
    try: 
        name = ta.Function(indicator)(data_full['c'])
        if len(name)>100:
            data_full[indicator] = name
        else:
            for i in range(len(name)):
                data_full[indicator] = name[i]
    except Exception as e:
        print(e)
        # print(e)

    # indi need 2 value
    try: 
        name = ta.Function(indicator)(data_full['h'],data_full['l'])
        if len(name)>100:
            data_full[indicator] = name
        else:
            for i in range(len(name)):
                data_full[indicator] = name[i]
    except Exception as e:
        print(e) 

    # indi need 3 value       
    try:
        name = ta.Function(indicator)(data_full['h'],data_full['l'],data_full['c'])
        if len(name)>100:
            data_full[indicator] = name
        else:
            for i in range(len(name)):
                data_full[indicator] = name[i]
    except Exception as e:
        print(e)

    # indi need 4 value
    try: 
        name = ta.Function(indicator)(data_full['o'],data_full['h'],data_full['l'],data_full['c'])
        if len(name)>100:
            data_full[indicator] = name
        else:
            for i in range(len(name)):
                data_full[indicator] = name[i]
    except Exception as e:
        print(e)

# others indi that need specific value
for hlcv_indicator in hlcv_indicators:
    try: 
        name = ta.Function(hlcv_indicator)(data_full['o'],data_full['h'],data_full['c'],data_full['vol'])
        if len(name)>100:
            data_full[indicator] = name
        else:
            for i in range(len(name)):
                data_full[indicator] = name[i]
    except Exception as e:
        print(e) 
for cl_indicator in cl_indicators:
    try: 
        name = ta.Function(cl_indicator)(data_full['c'],data_full['vol'])
        if len(name)>100:
            data_full[indicator] = name
        else:
            for i in range(len(name)):
                data_full[indicator] = name[i]
    except Exception as e:
        print(e) 
data_full_with_indi.append(data_full)


data_full_full = []
```
