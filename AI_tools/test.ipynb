```python
import pandas as pd
import talib
```

```python
data_full = pd.read_csv('Dataset/test.csv')
```

```python
import ta
df = ta.add_all_ta_features(data_full, "o", "h", "l", "c", "vol", fillna=True)
```

```python
df.dropna(inplace=True)
```

```python
df.drop(['confirm'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Min Max Scaler
# MinMaxScaler scales all the data features in the range [0, 1] or else in the range [-1, 1] if there are negative values in the dataset.

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df_normalized.describe()
```

```python
# Standard Scaler
# StandardScaler removes the mean and scales the data to unit variance.

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_zscore_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df_zscore_normalized.describe()
```

```python
# Robust Scaler
# RobustScaler removes the median and scales the data according to the quantile range (defaults to IQR: Interquartile Range).


from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
df_robust_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)   
df_robust_normalized.describe()
```

```python
# Binary Scaler
# Binarizer is used to convert the numerical features into binary values.

from sklearn.preprocessing import Binarizer

scaler = Binarizer(threshold=0.0)
df_binary_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df_binary_normalized.describe()
```
